# Matplotlib Pyplot

```python
import matplotlib.pyplot as plt
import numpy as np
```

## Simple plot

```python
plt.plot([1, 2, 3], [100, 200, 300])
```

```python
PI = np.pi
x = np.array([0, PI/6, PI/4, PI/3, PI/2])
y = np.sin(x)
```

```python
z = [(a, b) for a, b in zip(x, y)]
print(z)
```

```python
plt.plot(x, y)
```

```python
x = np.random.uniform(-1.0, 0.0, 10000)

# Plotting Histogram
plt.hist(x, 100)
```

## Plotting for arcsinhx

```python
x = np.array(range(100))
y = np.arcsinh(x)

plt.plot(x, y)
```

## Plotting for arccoshx

```python
x = np.array(range(1, 100))
y = np.arccosh(x)

plt.plot(x, y)
```

## Piechart

```python
labels = 'Cheetah', 'Lion', 'Monkey', 'Donkey'
sizes = [15, 30, 45, 10]
explode = (0, 0.1, 0, 0)  # only "explode" the 2nd slice (i.e. 'Hogs')

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
```
